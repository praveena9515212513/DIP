```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import *
img = plt.imread('bts.jpg')
plt.rcParams['figure.figsize']=(20,10)
plt.imshow(img)
plt.axis('off')
```

```python
img1 = np.rot90(img,1)
img2 = np.rot90(img,2)
img3 = np.rot90(img,3)
img4 = np.rot90(img,4)
img5 = np.fliplr(img)
img6 = np.flipud(img)
plt.subplot(231),plt.imshow(img1)
plt.subplot(232),plt.imshow(img2)
plt.subplot(233),plt.imshow(img3)
plt.subplot(234),plt.imshow(img4)
plt.subplot(235),plt.imshow(img5)
plt.subplot(236),plt.imshow(img6)
```

```python
def avg_gray_scale(img):
#     img = np.dot(img[:,:,:3],[1/3,1/3,1/3])
    rows,cols,temp = img.shape
    gs = np.zeros((rows,cols))
    for i in range(rows):
        for j in range(cols):
            gs[i,j]=np.average(img[i,j])
    return gs
def weight_avg_gray_scale(img):
    img = np.dot(img[:,:,:3],[0.299,0.587,0.114])
    return img
plt.subplot(121),plt.imshow(avg_gray_scale(img),cmap='gray')
plt.subplot(122),plt.imshow(weight_avg_gray_scale(img),cmap='gray')
```

```python
img_gray = avg_gray_scale(img)

def gray2bw(img,tresh):
    rows,cols = img.shape
    bw = np.zeros((rows,cols))
    for i in range(rows):
        for j in range(cols):
            bw[i,j]=0 if img[i,j]<=tresh else 255
    return bw
img_bw = gray2bw(img_gray,127)
plt.subplot(121),plt.imshow(img_gray,cmap='gray')
plt.subplot(122),plt.imshow(img_bw,cmap='gray')
```

```python
# convert from png to jpeg
def png2jpg(img):
    img = img[:,:,:3]
    img = img*255
    return np.uint8(np.clip(img.round(),0,255))
img = plt.imread('stickanimalsRGBA.png')
img_jpg = png2jpg(img)
print(img.shape)
print(img_jpg.shape)
plt.subplot(121),plt.imshow(img)
plt.subplot(122),plt.imshow(img_jpg)
```
